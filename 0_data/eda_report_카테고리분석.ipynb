# NutriFit 성분 수작업 분류 샘플 데이터 카테고리 분석

**이 분석은 전체 27,779건 중 수작업으로 성분을 확인한 45건 중, 카테고리 분류가 완료된 25건만을 대상으로 합니다.**

### 3. 카테고리별(멀티비타민/비타민C/오메가3/유산균/콜라겐/마그네슘) 상품 개수 집계
**이 분석을 왜 하는지**: 수작업으로 1차 분류한 샘플 데이터에서 핵심 성분 카테고리별 제품 분포가 어떻게 구성되어 있는지 파악하기 위함입니다.

```python
import pandas as pd

# 엑셀 로드 및 필터링
df_excel = pd.read_excel('성분_수작업입력_초안.xlsx', skiprows=3)
df_filtered = df_excel[df_excel['카테고리'].notna() & (df_excel['카테고리'] != '확인필요')]

print('=== 카테고리별 상품 개수 집계 ===')
category_counts = df_filtered['카테고리'].value_counts()
print(category_counts.to_string())
```

### 4. 플랫폼 x 카테고리 교차표(피벗 테이블, 합계 포함)
**이 분석을 왜 하는지**: 1차 분류 완료된 제품들이 플랫폼별로 각 카테고리에 어떻게 교차 분포하고 있는지 파악하여 플랫폼별 강점 카테고리를 비교하기 위함입니다.

```python
# 교차표 생성
pivot_table = pd.crosstab(df_filtered['플랫폼'], df_filtered['카테고리'], margins=True, margins_name='합계')
print('=== 플랫폼별 x 카테고리별 교차표 ===')
print(pivot_table.to_string())
```
